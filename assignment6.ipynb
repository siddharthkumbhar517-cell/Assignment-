 ➊ Given the temperature of a day in Degrees Celsius, convert this given 
 temperature from Celsius to Fahrenheit. Complete the function to do so. 
 Round the output up to 2 decimal places

```python
print("Question 1 solution:")
def celsius_farhen(Celsius):
    ans = (Celsius* 9/5)+32
    return round(ans,2)


tempInC = int(input("Enter temparature of the day in Celcius:"))
print("Temperature of Day in Celcius:",tempInC)
print(f"Temperature in Fahrenheit is: {celsius_farhen(tempInC)}")
```

 ➋ Write a program to calculate the total amount if the principal 
 amount, simple interest rate per annum, and time in the number of days 
 are given. The simple interest rate per annum is provided as a percentage. 
 Assuming that there are 365 days in a year, print the total amount rounded 
 up to 2 decimal places.
 Note 1 : Use round(ans,2) to round up ans up to two decimals

```python
print("Question 2 solution:")
def totalSimpleAmount(t, pa, ir):
    # t: time(in no. of days), pa: principal amt, ir: interest rate.
    si = (pa * ir * t)/(365*100) #simple interest
    amt = si + pa #total amount
    return print(f"Total amount is: Rs.{round(amt, 2)}\nFor PA: Rs.{pa}, on IR: {ir}% for time: {t} days.")

print("Total amount based on T, PA and IR.")
time = int(input("Enter the time in days:"))
pa = int(input("Enter the Pricipal Amount:"))
ir = int(input("Enter Interest Rate:"))

totalSimpleAmount(time, pa, ir)
```

 ➌ Write a program that takes the cost price of a vehicle as an argument and 
 returns the road tax, that the vehicle owner has to pay according to the 
 following criteria:
 Note1 : Return the tax up to one decimal place.

```python
def road_tax(price):
    if price > 100000:
        tax = 0.2 * price
    elif (price > 75000 and price <= 100000):
        tax = 0.15 * price
    elif (price > 50000 and price <= 75000):
        tax = 0.1 * price
    else:
        tax = 0.05 * price
    return print(f"Tax for: RS. {price} is: RS.{round(tax,1)}")

cp = int(input("Enter the Cost Price:"))
road_tax(cp)
```

 ➍ You are given a positive integer A denoting the radius of a circle. You have to calculate the area of the circle. Note: The formula for the area of a circle is Area = πr², where r is the circle's radius. You can use the value of PI as 3.1416. Round up the final answer up to 2 decimal places. You can round(area, 2) for rounding area to 2 decimal places


```python
class Solution:
    # @param A : integer
    # @return an integer
    def solve(self, A):
        # return math.ceil(math.pi*(A**2))
        return round(3.1416*(A**2), 2)
print("Question 4 solution:")
s = Solution()
r = int(input("Enter Radius of the circle:"))
print(f"Area of the circle for radius {r} is: {s.solve(r)}")
```

 ➎ Given an integer n, write a program to return the sum of squares of first n natural numbers in integer format.
 Note1 :
 The formula for the sum of squares of first n natural numbers is: n(n+1)(2n+1)/6

```python
def sum_squares(n):
    ans = (n*(n+1)*(2*n+1))//6
    # ''' input:Given Integer n
    #      output: Return ans as sum of squares in integer forma '''
    return print(f"Sum of squares for num: {n} is: {ans}")
print("Question 5 solution:")
n = int(input("Enter a number:"))
sum_squares(n)
```

 ➐ Given the height (A) and weight (B) of a person as input in centimetres and 
 kilograms.
 Find the BMI of that person and the classification of the user based on their 
 BMI.
 ➊ Print Underweight if BMI < 18.5
 ➋ Print Normal if BMI lies in the range (18.5, 24.9)
 ➌ Print Overweight if BMI lies in the range (24.9, 29.9)
 ➍ Print Obese if BMI is greater than 29.9
 If x is the weight in kilograms and y is the height in metres.
 Then, BMI is calculated as x/(y*y).
 Note: Use round(BMI,1) to get only one digit after decimal
 Input Format:
 The first line contains the height of the person in centimetres.
 The second line of the input contains the weight of the person in kilograms.
 Output Format:
 The first line of the output contains the classification of a person based on the BMI value.
 The second line of the output contains the BMI value of the person having only one digit after decimal

```python
def main(height_cm,weight):
    height = height_cm/100
    BMI = (weight/(height * height))
    # YOUR CODE GOES HERE
    # Please take input and print output to standard input/output
    return BMI
def classify(BMI):
    if BMI < 18.5:
        return "Underweight"
    elif 18.5 <= BMI <= 24.9:
        return "Normal"
    elif 24.9 < BMI <= 29.9:
        return "Overweight"
    else:
        return ("Obese")

print("Question 7 solution:")

height_in_cm = int(input("Enter your height in CM: "))
weight_in_kg = int(input("Enter your weight in KG: "))
print(f"BMI for H: {height_in_cm} CM & W: {weight_in_kg} KG")
BMI =  main(height_in_cm,weight_in_kg)
classification = classify(BMI)
rounded_bmi = round(BMI,1)
print("Classification: ",classification)
print("BMI: ",rounded_bmi)
```
